This notebook provides a initial overview over the values and correlation of already preprocessed metrics for sequential and concurrent experiments.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import sys
sys.path.append('../../src')
import global_config
from data import reward
```

```python
seq_metrics_df = reward._generate_unified_metrics_dataframe(global_config.HOSTS, True)
con_metrics_df = reward._generate_unified_metrics_dataframe(global_config.HOSTS, True)
```

```python
seq_metrics_df.boxplot(column=list(seq_metrics_df.columns.values),figsize=(70,10))
```

```python
con_metrics_df.boxplot(column=list(seq_metrics_df.columns.values),figsize=(70,10))
```

```python
def display_correlation(metrics_df):
    correlation_pearson = metrics_df.corr(method='pearson').fillna(0.0)
    np.fill_diagonal(correlation_pearson.values, 0.0)
    display(correlation_pearson.style.applymap(lambda value: 'color: green' if value > 0.75 else ('color: red' if value < -0.25 else 'color: black')))
```

```python
display('Correlation of sequential metrics for whole experiment')
display_correlation(seq_metrics_df)
```

```python
display('Correlation of concurrent metrics for whole experiment')
display_correlation(con_metrics_df)
```
