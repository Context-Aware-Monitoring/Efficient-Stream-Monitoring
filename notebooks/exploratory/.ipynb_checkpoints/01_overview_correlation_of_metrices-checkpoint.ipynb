```python
import pandas as pd
```

```python
sequential = True
```

```python
relevant_columns = ['now', 'cpu.user', 'mem.used', 'load.min1', 'load.min5', 'load.min15']
if sequential:
    wally113 = pd.read_csv('../../data/raw/sequential_data/metrics/wally113_metrics.csv').loc[:,relevant_columns]
    wally117 = pd.read_csv('../../data/raw/sequential_data/metrics/wally117_metrics.csv').loc[:,relevant_columns]
    wally122 = pd.read_csv('../../data/raw/sequential_data/metrics/wally122_metrics.csv').loc[:,relevant_columns]
    wally123 = pd.read_csv('../../data/raw/sequential_data/metrics/wally123_metrics.csv').loc[:,relevant_columns]
    wally124 = pd.read_csv('../../data/raw/sequential_data/metrics/wally124_metrics.csv').loc[:,relevant_columns]
else:    
    wally113 = pd.read_csv('../../data/raw/concurrent_data/metrics/wally113_metrics_concurrent.csv').loc[:,relevant_columns]
    wally117 = pd.read_csv('../../data/raw/concurrent_data/metrics/wally117_metrics_concurrent.csv').loc[:,relevant_columns]
    wally122 = pd.read_csv('../../data/raw/concurrent_data/metrics/wally122_metrics_concurrent.csv').loc[:,relevant_columns]
    wally123 = pd.read_csv('../../data/raw/concurrent_data/metrics/wally123_metrics_concurrent.csv').loc[:,relevant_columns]
    wally124 = pd.read_csv('../../data/raw/concurrent_data/metrics/wally124_metrics_concurrent.csv').loc[:,relevant_columns]
```

```python
def add_suffixes_except_now_column(df, name):
    df.columns=df.columns.map(lambda x : x+name if x !='now' else x)
```

```python
add_suffixes_except_now_column(wally113, '_wally113')
add_suffixes_except_now_column(wally117, '_wally117')
add_suffixes_except_now_column(wally122, '_wally122')
add_suffixes_except_now_column(wally123, '_wally123')
add_suffixes_except_now_column(wally124, '_wally124')
overall_df = wally113.merge(wally117, left_on='now', right_on='now')
overall_df = overall_df.merge(wally122, left_on='now', right_on='now')
overall_df = overall_df.merge(wally123, left_on='now', right_on='now')
overall_df = overall_df.merge(wally124, left_on='now', right_on='now')
overall_df.shape
```

```python
def highlight(s):
    threshold = 0.85
    threshold_negative = -0.25
    color = 'black'
    if s < threshold_negative:
        color = 'red'
    elif s > threshold:
        color = 'green'
    return 'color: %s' % color
```

```python
correlation_pearson = overall_df.corr(method='pearson')
correlation_pearson.style.applymap(highlight)
```

```python
def correlation_for_time_window(start, end, df):
    return df.loc[(df['now']>start) & (df['now']<end)].corr(method='pearson').style.applymap(highlight)
```

```python
correlation_for_time_window('2019-11-19 17:00:00 CEST', '2019-11-21 18:00:00 CEST', overall_df)
```
