```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
relevant_columns = ['now', 'cpu.user', 'mem.used', 'load.min1', 'load.min5', 'load.min15']

wally113 = pd.read_csv('../../data/raw/sequential_data/metrics/wally113_metrics.csv').loc[:,relevant_columns]
wally117 = pd.read_csv('../../data/raw/sequential_data/metrics/wally117_metrics.csv').loc[:,relevant_columns]
wally122 = pd.read_csv('../../data/raw/sequential_data/metrics/wally122_metrics.csv').loc[:,relevant_columns]
wally123 = pd.read_csv('../../data/raw/sequential_data/metrics/wally123_metrics.csv').loc[:,relevant_columns]
wally124 = pd.read_csv('../../data/raw/sequential_data/metrics/wally124_metrics.csv').loc[:,relevant_columns]
data = [wally113, wally117, wally122, wally123, wally124]
```

```python
def collect_columns_from_data(name):
    return {'wally113': wally113[name], 'wally117': wally117[name], 'wally122': wally122[name], 'wally123': wally123[name], 'wally124': wally124[name]}

def boxplot_column(name, ax, c=['wally113', 'wally117', 'wally122', 'wally123', 'wally124']):
    data = collect_columns_from_data(name)
    data_df = pd.DataFrame(data)
    data_df.boxplot(column=c, ax=ax,showfliers=False)
    
def plot_load_metric(name, sequential=True):
    fig, ax = plt.subplots(1, 2)
    fig.suptitle('%s %s' % (name, 'sequential' if sequential else 'concurrent'))
    yl = '%s in percent' % name
    ax[0].set(xlabel='node', ylabel=yl, title='control node')
    ax[1].set(xlabel='node', ylabel=yl, title='computing nodes')
    boxplot_column(name, ax[0], c=['wally113'])
    boxplot_column(name, ax[1], c=['wally117', 'wally122', 'wally123', 'wally124'])    
```

```python
fig, ax = plt.subplots()
ax.set(xlabel='node', ylabel='cpu.user in %', title='cpu.user metric sequential')
boxplot_column('cpu.user', ax)
```

```python
fig, ax = plt.subplots()
ax.set(xlabel='node', ylabel='mem.user in %', title='mem.user metric sequential')
boxplot_column('mem.used', ax)
```

```python
plot_load_metric('load.min1')
```

```python
plot_load_metric('load.min5')
```

```python
plot_load_metric('load.min15')
```
